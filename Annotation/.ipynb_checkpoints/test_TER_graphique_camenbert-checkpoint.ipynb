```python
# Installation des dépendances (si nécessaire, peuvent être commentées après la première exécution)
!pip install datasets transformers scikit-learn accelerate torch emoji matplotlib -q # Ajout de matplotlib

import pandas as pd
from datasets import Dataset
from transformers import AutoTokenizer, AutoModelForSequenceClassification, TrainingArguments, Trainer, EarlyStoppingCallback
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, f1_score
from sklearn.model_selection import train_test_split
import torch
import gc
from accelerate import Accelerator
import re
import nltk # Peut être optionnel si pas de tokenisation NLTK active
import matplotlib.pyplot as plt # Ajouté pour le camembert
import numpy as np # Souvent utile avec matplotlib
from google.colab import drive # Import colab drive
from IPython.display import display # Pour un meilleur affichage des DataFrames

# Initialiser l'accélérateur explicitement
accelerator = Accelerator()

# --- SECTION ADAPTÉE DE VOTRE SCRIPT D'ANALYSE : CHARGEMENT ET PRÉPARATION DES DONNÉES ---
drive.mount('/content/gdrive/', force_remount=True)
notebook_path = '/content/gdrive/My Drive/Colab Notebooks/' # Ajustez si votre chemin est différent
%cd "$notebook_path"

print("--- CHARGEMENT ET PRÉPARATION DES DONNÉES (À PARTIR DU .TSV) ---")
file_path_final_annotations = 'Annotation_apres_mise_en_accord_a_utiliser_pour_entrainement_modele.tsv'
try:
    df_full = pd.read_csv(file_path_final_annotations, sep='\t')
except FileNotFoundError:
    print(f"ERREUR: Le fichier '{file_path_final_annotations}' est introuvable.")
    raise
print(f"Nombre total de tweets chargés : {len(df_full)}")

label_column_original = 'emotion_principale'
text_column_original = 'text'

if label_column_original not in df_full.columns:
    raise ValueError(f"Colonne des labels d'origine '{label_column_original}' introuvable.")
if text_column_original not in df_full.columns:
    raise ValueError(f"Colonne de texte d'origine '{text_column_original}' introuvable.")

def clean_text_for_transformer(text):
    if not isinstance(text, str): return ""
    text = text.lower()
    text = re.sub(r"http\S+|www\S+", "[URL]", text)
    text = re.sub(r"@\w+", "[USER]", text)
    text = re.sub(r"#\w+", "[HASHTAG]", text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

print("Application du nettoyage de texte...")
df_full['cleaned_text'] = df_full[text_column_original].apply(clean_text_for_transformer)

votre_label_map_nom = {
    1: 'Peur', 2: 'Colère', 3: 'Joie', 4: 'Surprise',
    5: 'Tristesse', 6: 'Dégoût', 7: 'Neutre'
}

model_specific_label_mapping = {
    1: 2, 2: 0, 3: 3, 4: 6, 5: 5, 6: 1, 7: 4
}
labels_du_modele_distilroberta = ["anger", "disgust", "fear", "joy", "neutral", "sadness", "surprise"]

df_full['model_label'] = df_full[label_column_original].map(model_specific_label_mapping)
df_full.dropna(subset=['model_label'], inplace=True)
df_full['model_label'] = df_full['model_label'].astype(int)

print("Division des données en ensembles d'entraînement et de test...")
train_df, test_df = train_test_split(
    df_full[['cleaned_text', 'model_label', label_column_original, text_column_original]],
    test_size=0.2,
    random_state=42,
    stratify=df_full['model_label']
)

train_df_hf = train_df[['cleaned_text', 'model_label']].rename(columns={'cleaned_text': 'text', 'model_label': 'label'})
test_df_hf = test_df[['cleaned_text', 'model_label']].rename(columns={'cleaned_text': 'text', 'model_label': 'label'})

print(f"Taille de l'ensemble d'entraînement (HF): {len(train_df_hf)}")
print(f"Taille de l'ensemble de test (HF): {len(test_df_hf)}")

print("Conversion en datasets Hugging Face...")
train_dataset = Dataset.from_pandas(train_df_hf)
test_dataset_for_eval = Dataset.from_pandas(test_df_hf)

del df_full
del train_df_hf
del test_df_hf
gc.collect()


print("Chargement du modèle et du tokenizer...")
model_name = "j-hartmann/emotion-english-distilroberta-base"
tokenizer = AutoTokenizer.from_pretrained(model_name)
num_labels_model = 7
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=num_labels_model)

def tokenize_function(examples):
    return tokenizer(examples["text"], padding="max_length", truncation=True, max_length=64)

print("Tokenisation des données...")
tokenized_train = train_dataset.map(tokenize_function, batched=True, remove_columns=["text"])
tokenized_test_for_eval = test_dataset_for_eval.map(tokenize_function, batched=True, remove_columns=["text"])

del train_dataset
del test_dataset_for_eval
gc.collect()

def compute_metrics(pred):
    labels = pred.label_ids
    preds = pred.predictions.argmax(-1)
    precision, recall, f1, _ = precision_recall_fscore_support(labels, preds, average='weighted', zero_division=0)
    acc = accuracy_score(labels, preds)
    class_f1_scores = f1_score(labels, preds, average=None, zero_division=0, labels=list(range(num_labels_model)))
    class_metrics = {f"f1_class_{i} ({labels_du_modele_distilroberta[i]})": score for i, score in enumerate(class_f1_scores)}
    return {'accuracy': acc, 'f1': f1, 'precision': precision, 'recall': recall, **class_metrics}

print("Configuration de l'entraînement...")
training_args = TrainingArguments(
    output_dir="./results_distilroberta_from_tsv_compare", # Changer le nom pour cette version
    learning_rate=2e-5,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=16,
    num_train_epochs=3,
    weight_decay=0.01,
    eval_strategy="steps",
    eval_steps=50,
    save_strategy="steps",
    save_steps=50,
    load_best_model_at_end=True,
    metric_for_best_model="f1",
    push_to_hub=False,
    report_to="none",
    logging_strategy="steps",
    logging_steps=10,
    fp16=torch.cuda.is_available(),
)

print("Initialisation du Trainer...")
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_train,
    eval_dataset=tokenized_test_for_eval,
    tokenizer=tokenizer,
    compute_metrics=compute_metrics,
    callbacks=[EarlyStoppingCallback(early_stopping_patience=3)]
)

print("Début de l'entraînement...")
try:
    trainer.train()
except RuntimeError as e:
    if "out of memory" in str(e).lower():
        print("Erreur CUDA Out of Memory détectée.")
        if torch.cuda.is_available():
            torch.cuda.empty_cache()
    else:
        raise e

model_path_saved = training_args.output_dir

if hasattr(trainer, 'state') and trainer.state.best_model_checkpoint is not None:
    print("Évaluation du meilleur modèle sur l'ensemble de test...")
    eval_results = trainer.evaluate(eval_dataset=tokenized_test_for_eval)
    print(f"Résultats d'évaluation: {eval_results}")
    trainer.save_model(model_path_saved)
    print(f"Meilleur modèle sauvegardé à {model_path_saved}")
else:
    print("L'entraînement n'a pas pu se terminer ou aucun meilleur modèle n'a été trouvé.")
    try:
        model.save_pretrained(model_path_saved)
        tokenizer.save_pretrained(model_path_saved)
        print(f"Modèle actuel (potentiellement non optimal) sauvegardé à {model_path_saved}")
    except Exception as e_save:
        print(f"Impossible de sauvegarder le modèle actuel: {e_save}")

del model
del tokenized_train
del tokenized_test_for_eval
if 'trainer' in locals(): del trainer
if torch.cuda.is_available():
    torch.cuda.empty_cache()
gc.collect()


# --- SECTION DE PRÉDICTION SUR L'ENSEMBLE DU JEU DE TEST ---
print("\n--- PRÉDICTIONS SUR L'ENSEMBLE DU JEU DE TEST ---")
all_predictions_data = []

try:
    reloaded_model = AutoModelForSequenceClassification.from_pretrained(model_path_saved)
    reloaded_tokenizer = AutoTokenizer.from_pretrained(model_path_saved)
    if torch.cuda.is_available():
        reloaded_model.to(accelerator.device)
    reloaded_model.eval()

    for index, row in test_df.iterrows():
        text_to_predict = row["cleaned_text"]
        true_original_label_numeric = row[label_column_original] # Votre label 1-7
        # true_model_label_numeric = row["model_label"] # Label 0-6 utilisé pour l'entraînement (vérité terrain modèle)

        inputs = reloaded_tokenizer(text_to_predict, return_tensors="pt", padding=True, truncation=True, max_length=64)
        if torch.cuda.is_available():
            inputs = {k: v.to(reloaded_model.device) for k, v in inputs.items()}

        with torch.no_grad():
            outputs = reloaded_model(**inputs)

        probs = torch.nn.functional.softmax(outputs.logits, dim=-1)
        predicted_class_index = outputs.logits.argmax(-1).item() # Index 0-6

        true_emotion_original_name = votre_label_map_nom.get(true_original_label_numeric, "Inconnu (original)")
        predicted_emotion_name = labels_du_modele_distilroberta[predicted_class_index]

        prediction_entry = {
            "emotion_reelle_originale_nom": true_emotion_original_name, # Basé sur votre mapping 1-7
            "emotion_predite_nom": predicted_emotion_name, # Nom basé sur le mapping du modèle 0-6
        }
        all_predictions_data.append(prediction_entry)

    df_predictions = pd.DataFrame(all_predictions_data)
    print("\nDataFrame des prédictions (émotions réelles vs prédites) sur le jeu de test:")
    display(df_predictions.head())


    # --- SECTION POUR LES DIAGRAMMES EN CAMEMBERT COMPARATIFS ---
    print("\n--- DIAGRAMMES EN CAMEMBERT : ANNOTATIONS RÉELLES VS PRÉDITES (JEU DE TEST) ---")

    if not df_predictions.empty:
        # 1. Camembert des annotations réelles (basé sur 'emotion_reelle_originale_nom')
        real_emotion_counts = df_predictions["emotion_reelle_originale_nom"].value_counts()
        if not real_emotion_counts.empty:
            fig_real, ax_real = plt.subplots(figsize=(8,7)) # Ajuster la taille
            ax_real.pie(real_emotion_counts.values, labels=real_emotion_counts.index, autopct='%1.1f%%', startangle=90)
            ax_real.axis('equal')
            plt.title("Distribution des annotations réelles sur le jeu de test")
            plt.show()
        else:
            print("Aucune annotation réelle trouvée dans df_predictions.")

        # 2. Camembert des annotations prédites (basé sur 'emotion_predite_nom')
        predicted_emotion_counts = df_predictions["emotion_predite_nom"].value_counts()
        if not predicted_emotion_counts.empty:
            fig_pred, ax_pred = plt.subplots(figsize=(8,7)) # Ajuster la taille
            ax_pred.pie(predicted_emotion_counts.values, labels=predicted_emotion_counts.index, autopct='%1.1f%%', startangle=90)
            ax_pred.axis('equal')
            plt.title("Distribution des annotations prédites par le modèle sur le jeu de test")
            plt.show()
        else:
            print("Aucune annotation prédite trouvée dans df_predictions.")
    else:
        print("Aucune prédiction n'a été générée pour créer les camemberts.")


except NameError:
    print(f"Variable 'model_path_saved' non définie ou l'entraînement a échoué. Le modèle n'a probablement pas été sauvegardé, section de prédiction annulée.")
except OSError:
    print(f"Le modèle n'a pas été trouvé à '{model_path_saved}'. Section de prédiction annulée.")
    print("Vérifiez que l'entraînement s'est bien terminé et que le chemin de sauvegarde est correct.")
except Exception as e:
    print(f"Erreur lors de la section de prédiction et de visualisation: {e}")
    import traceback
    traceback.print_exc()


print("\nScript terminé.")
```

```python
# Installation des dépendances (si nécessaire, peuvent être commentées après la première exécution)
!pip install datasets transformers scikit-learn accelerate torch emoji matplotlib -q # Ajout de matplotlib

import pandas as pd
from datasets import Dataset
from transformers import AutoTokenizer, AutoModelForSequenceClassification, TrainingArguments, Trainer, EarlyStoppingCallback
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, f1_score
from sklearn.model_selection import train_test_split
import torch
import gc
from accelerate import Accelerator
import re
import nltk # Peut être optionnel si pas de tokenisation NLTK active
import matplotlib.pyplot as plt # Ajouté pour le camembert
import numpy as np # Souvent utile avec matplotlib
from google.colab import drive # Import colab drive
from IPython.display import display # Pour un meilleur affichage des DataFrames

# Initialiser l'accélérateur explicitement
accelerator = Accelerator()

# --- SECTION ADAPTÉE DE VOTRE SCRIPT D'ANALYSE : CHARGEMENT ET PRÉPARATION DES DONNÉES ---
drive.mount('/content/gdrive/', force_remount=True)
notebook_path = '/content/gdrive/My Drive/Colab Notebooks/' # Ajustez si votre chemin est différent
%cd "$notebook_path"

print("--- CHARGEMENT ET PRÉPARATION DES DONNÉES (À PARTIR DU .TSV) ---")
file_path_final_annotations = 'Annotation_apres_mise_en_accord_a_utiliser_pour_entrainement_modele.tsv'
try:
    df_full = pd.read_csv(file_path_final_annotations, sep='\t')
except FileNotFoundError:
    print(f"ERREUR: Le fichier '{file_path_final_annotations}' est introuvable.")
    raise
print(f"Nombre total de tweets chargés : {len(df_full)}")

label_column_original = 'emotion_principale'
text_column_original = 'text'

if label_column_original not in df_full.columns:
    raise ValueError(f"Colonne des labels d'origine '{label_column_original}' introuvable.")
if text_column_original not in df_full.columns:
    raise ValueError(f"Colonne de texte d'origine '{text_column_original}' introuvable.")

def clean_text_for_transformer(text):
    if not isinstance(text, str): return ""
    text = text.lower()
    text = re.sub(r"http\S+|www\S+", "[URL]", text)
    text = re.sub(r"@\w+", "[USER]", text)
    text = re.sub(r"#\w+", "[HASHTAG]", text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

print("Application du nettoyage de texte...")
df_full['cleaned_text'] = df_full[text_column_original].apply(clean_text_for_transformer)

votre_label_map_nom = {
    1: 'Peur', 2: 'Colère', 3: 'Joie', 4: 'Surprise',
    5: 'Tristesse', 6: 'Dégoût', 7: 'Neutre'
}

model_specific_label_mapping = {
    1: 2, 2: 0, 3: 3, 4: 6, 5: 5, 6: 1, 7: 4 # Peur(1)->fear(2), Colère(2)->anger(0), etc.
}
labels_du_modele_distilroberta = ["anger", "disgust", "fear", "joy", "neutral", "sadness", "surprise"]

df_full['model_label'] = df_full[label_column_original].map(model_specific_label_mapping)
df_full.dropna(subset=['model_label'], inplace=True)
df_full['model_label'] = df_full['model_label'].astype(int)

print("Division des données en ensembles d'entraînement et de test...")
train_df, test_df = train_test_split(
    df_full[['cleaned_text', 'model_label', label_column_original, text_column_original]],
    test_size=0.2,
    random_state=42,
    stratify=df_full['model_label']
)

train_df_hf = train_df[['cleaned_text', 'model_label']].rename(columns={'cleaned_text': 'text', 'model_label': 'label'})
test_df_hf = test_df[['cleaned_text', 'model_label']].rename(columns={'cleaned_text': 'text', 'model_label': 'label'})

print(f"Taille de l'ensemble d'entraînement (HF): {len(train_df_hf)}")
print(f"Taille de l'ensemble de test (HF): {len(test_df_hf)}")

print("Conversion en datasets Hugging Face...")
train_dataset = Dataset.from_pandas(train_df_hf)
test_dataset_for_eval = Dataset.from_pandas(test_df_hf)

del df_full
del train_df_hf
del test_df_hf # test_df est gardé pour l'itération finale
gc.collect()


print("Chargement du modèle et du tokenizer...")
model_name = "j-hartmann/emotion-english-distilroberta-base"
tokenizer = AutoTokenizer.from_pretrained(model_name)
num_labels_model = 7
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=num_labels_model)

def tokenize_function(examples):
    return tokenizer(examples["text"], padding="max_length", truncation=True, max_length=64)

print("Tokenisation des données...")
tokenized_train = train_dataset.map(tokenize_function, batched=True, remove_columns=["text"])
tokenized_test_for_eval = test_dataset_for_eval.map(tokenize_function, batched=True, remove_columns=["text"])

del train_dataset
del test_dataset_for_eval
gc.collect()

def compute_metrics(pred):
    labels = pred.label_ids
    preds = pred.predictions.argmax(-1)
    precision, recall, f1, _ = precision_recall_fscore_support(labels, preds, average='weighted', zero_division=0)
    acc = accuracy_score(labels, preds)
    class_f1_scores = f1_score(labels, preds, average=None, zero_division=0, labels=list(range(num_labels_model)))
    class_metrics = {f"f1_class_{i} ({labels_du_modele_distilroberta[i]})": score for i, score in enumerate(class_f1_scores)}
    return {'accuracy': acc, 'f1': f1, 'precision': precision, 'recall': recall, **class_metrics}

print("Configuration de l'entraînement...")
training_args = TrainingArguments(
    output_dir="./results_distilroberta_from_tsv_v2", # Changer le nom du dossier de sortie pour éviter conflits
    learning_rate=2e-5,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=16,
    num_train_epochs=3,
    weight_decay=0.01,
    eval_strategy="steps",
    eval_steps=50,
    save_strategy="steps",
    save_steps=50,
    load_best_model_at_end=True,
    metric_for_best_model="f1",
    push_to_hub=False,
    report_to="none",
    logging_strategy="steps",
    logging_steps=10,
    fp16=torch.cuda.is_available(),
)

print("Initialisation du Trainer...")
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_train,
    eval_dataset=tokenized_test_for_eval,
    tokenizer=tokenizer,
    compute_metrics=compute_metrics,
    callbacks=[EarlyStoppingCallback(early_stopping_patience=3)]
)

print("Début de l'entraînement...")
try:
    trainer.train()
except RuntimeError as e:
    if "out of memory" in str(e).lower():
        print("Erreur CUDA Out of Memory détectée.")
        if torch.cuda.is_available():
            torch.cuda.empty_cache()
    else:
        raise e

model_path_saved = training_args.output_dir # Utiliser le output_dir défini dans TrainingArguments

if hasattr(trainer, 'state') and trainer.state.best_model_checkpoint is not None:
    print("Évaluation du meilleur modèle sur l'ensemble de test...")
    eval_results = trainer.evaluate(eval_dataset=tokenized_test_for_eval)
    print(f"Résultats d'évaluation: {eval_results}")
    trainer.save_model(model_path_saved)
    print(f"Meilleur modèle sauvegardé à {model_path_saved}")
else:
    print("L'entraînement n'a pas pu se terminer ou aucun meilleur modèle n'a été trouvé.")
    try:
        model.save_pretrained(model_path_saved)
        tokenizer.save_pretrained(model_path_saved)
        print(f"Modèle actuel (potentiellement non optimal) sauvegardé à {model_path_saved}")
    except Exception as e_save:
        print(f"Impossible de sauvegarder le modèle actuel: {e_save}")

del model
del tokenized_train
del tokenized_test_for_eval
if 'trainer' in locals(): del trainer
if torch.cuda.is_available():
    torch.cuda.empty_cache()
gc.collect()


# --- SECTION DE PRÉDICTION SUR QUELQUES EXEMPLES DU JEU DE TEST AVEC CAMEMBERT PAR TWEET ---
print("\n--- EXEMPLES DE PRÉDICTIONS AVEC DISTRIBUTION DES PROBABILITÉS (CAMEMBERT) ---")

# Définir le nombre d'exemples à afficher
nombre_exemples_a_afficher = 5 # Vous pouvez changer ce nombre

try:
    reloaded_model = AutoModelForSequenceClassification.from_pretrained(model_path_saved)
    reloaded_tokenizer = AutoTokenizer.from_pretrained(model_path_saved)
    if torch.cuda.is_available():
        reloaded_model.to(accelerator.device)
    reloaded_model.eval()

    # Sélectionner quelques exemples aléatoires du jeu de test
    # S'assurer qu'il y a assez de données dans test_df
    if len(test_df) < nombre_exemples_a_afficher:
        print(f"Attention: Moins d'exemples disponibles ({len(test_df)}) que demandé ({nombre_exemples_a_afficher}). Affichage de tous les exemples disponibles.")
        sample_for_testing = test_df.copy()
    else:
        sample_for_testing = test_df.sample(n=nombre_exemples_a_afficher, random_state=42) # random_state pour reproductibilité

    for index, row in sample_for_testing.iterrows():
        text_to_predict = row["cleaned_text"]
        true_original_label_numeric = row[label_column_original]
        original_tweet_text = row[text_column_original]

        inputs = reloaded_tokenizer(text_to_predict, return_tensors="pt", padding=True, truncation=True, max_length=64)
        if torch.cuda.is_available():
            inputs = {k: v.to(reloaded_model.device) for k, v in inputs.items()}

        with torch.no_grad():
            outputs = reloaded_model(**inputs)

        probs_tensor = torch.nn.functional.softmax(outputs.logits, dim=-1)
        probabilities = probs_tensor[0].cpu().numpy() # Convertir en array NumPy
        predicted_class_index = probabilities.argmax()

        true_emotion_original_name = votre_label_map_nom.get(true_original_label_numeric, "Inconnu (original)")
        predicted_emotion_name = labels_du_modele_distilroberta[predicted_class_index]
        confidence_principale = probabilities[predicted_class_index]

        print(f"\n--- Exemple de Tweet (Index original: {index}) ---")
        print(f"Tweet original: {original_tweet_text}")
        # print(f"Tweet nettoyé: {text_to_predict}") # Optionnel
        print(f"Émotion réelle (votre annotation): {true_emotion_original_name} (Label original: {true_original_label_numeric})")
        print(f"Émotion prédite principale par le modèle: {predicted_emotion_name} (Confiance: {confidence_principale*100:.2f}%)")

        # Création du diagramme en camembert pour ce tweet
        fig, ax = plt.subplots(figsize=(8, 6)) # Ajuster la taille si besoin
        # Filtrer les labels et probabilités pour ne pas afficher ceux à 0% (ou très bas) pour plus de lisibilité
        threshold = 0.01 # Seuil de probabilité pour afficher (1%)
        filtered_labels = [label for i, label in enumerate(labels_du_modele_distilroberta) if probabilities[i] >= threshold]
        filtered_probs = [prob for prob in probabilities if prob >= threshold]

        if not filtered_labels: # Si toutes les probabilités sont sous le seuil, afficher la principale
             filtered_labels = [predicted_emotion_name]
             filtered_probs = [confidence_principale]


        wedges, texts, autotexts = ax.pie(filtered_probs, labels=filtered_labels, autopct='%1.1f%%', startangle=90)
        ax.axis('equal')
        plt.title(f"Distribution des probabilités d'émotion pour le tweet:\n'{original_tweet_text[:50]}...'") # Titre avec début du tweet

        # Améliorer la lisibilité des labels et pourcentages
        for text in texts:
            text.set_fontsize(9)
        for autotext in autotexts:
            autotext.set_fontsize(8)
            autotext.set_color('white')


        plt.tight_layout() # Ajuste automatiquement les paramètres du subplot pour donner un ajustement serré.
        plt.show()
        print("-" * 80)


except NameError:
    print(f"Variable 'model_path_saved' non définie ou l'entraînement a échoué. Le modèle n'a probablement pas été sauvegardé, section de prédiction annulée.")
except OSError:
    print(f"Le modèle n'a pas été trouvé à '{model_path_saved}'. Section de prédiction annulée.")
    print("Vérifiez que l'entraînement s'est bien terminé et que le chemin de sauvegarde est correct.")
except Exception as e:
    print(f"Erreur lors de la section de prédiction et de visualisation: {e}")
    import traceback
    traceback.print_exc()


print("\nScript terminé.")
```
